```python
import numpy as np
import cv2
import json
import os 
```

```python
JSON_PATH = 'project-18-at-2023-12-19-09-39-9f0466e1.json'
IMAGES_PATH = 'images'
LABELS_PATH = 'labels2'
LABELS_COLOR_PATH = 'labels2_color'

# TODO buat list di text aja
Fish = 0
Jallyfish = 1
Penguin = 2
Shark = 3
Starfish = 4
Stringray = 5

Fish_COLOR = (0,175,102)
Jellyfish_COLOR = (186,12,47)
Penguin_COLOR = (147,149,152)
Shark_COLOR = (23,28,143)
Starfish_COLOR = (239,149,207)
Stringray_COLOR = (78,54,152)


if not os.path.exists(LABELS_PATH): 
    os.makedirs(LABELS_PATH) 

if not os.path.exists(LABELS_COLOR_PATH): 
    os.makedirs(LABELS_COLOR_PATH) 
```

```python
f = open(JSON_PATH)
data = json.load(f)
for dt in data:
    file_upload = dt['file_upload']
    print(file_upload)
    img = cv2.imread(IMAGES_PATH+'\\'+file_upload)
    height, width, channels = img.shape
    label_image  = np.zeros([height, width], dtype=np.uint8)
    label_color  = np.zeros([height, width, 3], dtype=np.uint8)

    for ano in dt['annotations']:
        for res in ano['result']:
            ori_height = res['original_height']
            ori_width = res['original_width']
            if height!=ori_height and width!=ori_width:
              raise Exception("Size not match!")

            label = res['value']['polygonlabels'][0]
            labelnum = 0
            labelcol = (255,255,255)
            if label=='Fish':
                labelnum = Fish
                labelcol = Fish_COLOR
            elif label=='Jellyfish':
                labelnum = Jallyfish
                labelcol = Jellyfish_COLOR
            elif label=='Penguin':
                labelnum = Penguin
                labelcol = Penguin_COLOR
            elif label=='Shark':
                labelnum = Shark
                labelcol = Shark_COLOR
            elif label=='Starfish':
                labelnum = Starfish
                labelcol = Starfish_COLOR
            elif label=='Stringray':
                labelnum = Stringray
                labelcol = Stringray_COLOR
            print(label)

            points = res['value']['points']
            points_abs = [
                    (int(x / 100 * width), int(y / 100 * height)) for x, y in points
                ]
            points_np = np.array(points_abs)
            cv2.fillPoly(label_image, pts = [points_np], color =labelnum)
            cv2.fillPoly(label_color, pts = [points_np], color =labelcol)
    
    cv2.imwrite(LABELS_PATH+'\\'+file_upload+'.png', label_image) 
    cv2.imwrite(LABELS_COLOR_PATH+'\\'+file_upload+'.png', label_color) 


        
```
